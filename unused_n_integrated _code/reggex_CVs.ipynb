```python
import os
import fitz  # PyMuPDF
import pandas as pd
import re
```

# all CVs from a folder
returns a dataframe of all CVs with the information taken by reggex(Email, number, university)

```python
def extract_text_from_pdf(pdf_path):
    # """Extract text from a single PDF file."""
    with fitz.open(pdf_path) as doc:
        text = ""
        for page in doc:
            text += page.get_text()
    return text

def extract_emails_from_text(text):
    """Extract emails from text using regex."""
    email_pattern = r'[\w.+-]+@[\w-]+\.[a-zA-Z]{2,}'
    return re.findall(email_pattern, text)

def extract_phone_numbers_from_text(text):
    """Extract Saudi Arabian phone numbers from text using regex."""
    phone_pattern = r'\+966\s?[0-9]{1,3}\s?[0-9]{1,3}\s?[0-9]{1,4}|05\s?[0-9]{1,3}\s?[0-9]{1,3}\s?[0-9]{1,4}'
    return re.findall(phone_pattern, text)

def extract_university_names(text):
    """Extract University names from text using regex."""
    university_pattern = r'\b(?:[A-Z][a-z]*\s?)+\bUniversity'
    return re.findall(university_pattern, text)
    

def extract_texts_from_folder(folder_path):
    """Extract text from all PDF files in a folder and return as a Pandas DataFrame."""
    data = []
    for filename in os.listdir(folder_path):
        if filename.endswith(".pdf"):
            pdf_path = os.path.join(folder_path, filename)
            text = extract_text_from_pdf(pdf_path)
            emails = extract_emails_from_text(text)
            phone_numbers = extract_phone_numbers_from_text(text)
            universities = extract_university_names(text)
            data.append({"Filename": filename, "Text": text, "Emails": emails, "Phone Numbers": phone_numbers, "Universities": universities})
    
    return pd.DataFrame(data)

# Specify the folder containing the PDF files
folder_path = "CVs/"

# Extract texts and save as a DataFrame
pdf_texts_df = extract_texts_from_folder(folder_path)


# Display the DataFrame
pdf_texts_df
```

returns a dictionary of the data(only the first instance)

```python
import fitz  # PyMuPDF
import re

def extract_text_from_pdf(pdf_path):
    """Extract text from a single PDF file."""
    with fitz.open(pdf_path) as doc:
        text = ""
        for page in doc:
            text += page.get_text()
    return text

def extract_email_from_text(text):
    """Extract the first email from text using regex."""
    email_pattern = r'[\w.+-]+@[\w-]+\.[a-zA-Z]{2,}'
    match = re.search(email_pattern, text)
    return match.group(0) if match else None

def extract_phone_number_from_text(text):
    """Extract the first Saudi Arabian phone number from text using regex."""
    phone_pattern = r'\+966\s?[0-9]{1,3}\s?[0-9]{1,3}\s?[0-9]{1,4}|05\s?[0-9]{1,3}\s?[0-9]{1,3}\s?[0-9]{1,4}'
    match = re.search(phone_pattern, text)
    return match.group(0) if match else None

def extract_university_name(text):
    """Extract the first University name from text using regex."""
    university_pattern = r'\b(?:[A-Z][a-z]*\s?)+\bUniversity'
    match = re.search(university_pattern, text)
    return match.group(0) if match else None

def extract_info_from_pdf(pdf_path):
    """Extract information from a single PDF file and save it in a dictionary."""
    text = extract_text_from_pdf(pdf_path)
    email = extract_email_from_text(text)
    phone_number = extract_phone_number_from_text(text)
    university = extract_university_name(text)
    
    return {
        "Filename": pdf_path,
        "Text": text,
        "Email": email,
        "Phone Number": phone_number,
        "University": university
    }

# Specify the PDF file path
pdf_path = "CVs/Ali_Babonji_1.pdf"

# Extract information from the PDF file
pdf_info = extract_info_from_pdf(pdf_path)

# Display the dictionary
pdf_info
```
